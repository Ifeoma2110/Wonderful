```python
import pandas as pd
import numpy as np
```

```python
Location = "datasets/DC_Properties.csv"
df = pd. read_csv(Location)
df.head()
```

```python
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
result = smf.ols('PRICE ~ BATHRM + KITCHENS + BEDRM', data=df).fit()
result.summary()
```

```python
result = smf.ols('PRICE ~ ZIPCODE + KITCHENS + ROOMS + BEDRM + STORIES + BATHRM + HF_BATHRM + NUM_UNITS + FIREPLACES ', data=df).fit()
result.summary()
```

```python
#remove STORIES AND NUM_UNITS
```

```python
result = smf.ols('PRICE ~ ZIPCODE + KITCHENS + ROOMS + BEDRM + BATHRM + HF_BATHRM + FIREPLACES ', data=df).fit()
result.summary()
```

```python
result = smf.ols('PRICE ~ KITCHENS + ZIPCODE+ BEDRM + BATHRM + HF_BATHRM ', data=df).fit()
result.summary()
```
